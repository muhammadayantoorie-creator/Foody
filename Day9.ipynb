```python
# 1. Initialize a set
s = {1, 2, 3, 4, 5}
print(f"Initial set: {s}")

# 2. Add (adds a single element)
s.add(6)
print(f"After add(6): {s}")

# 3. Update (adds multiple elements at once)
s.update([7, 8, 9])
print(f"After update([7, 8, 9]): {s}")

# 4. Discard (removes an item; does nothing if the item isn't there)
s.discard(5)
print(f"After discard(5): {s}")

# 5. Remove (removes an item; will raise an error if the item isn't there)
s.remove(4)
print(f"After remove(4): {s}")

# 6. Pop (removes and returns a random element from the set)
popped_item = s.pop()
print(f"Popped item: {popped_item}")
print(f"Final set: {s}")
```

```python
#Set operation:
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union
print(A | B)
print(A.union(B))

# Intersection
print(A & B)
print(A.intersection(B))

# Difference
print(A - B)
print(A.difference(B))
print(B - A)

# Symmetric Difference (in either but not both)
print(A ^ B)
print(A.symmetric_difference(B))

# Subset/Superset
C = {1, 2}
print(C.issubset(A))
print(A.issuperset(C))
print(A.isdisjoint(B))
```

```python
set_a = {"flour", "sugar", "butter"}
set_b = {"sugar", "eggs"}
print(set_a.union(set_b))
print(set_a.intersection(set_b))
print(set_a.difference(set_b))

print(set_a | set_b)
print(set_a & set_b)
print(set_a - set_b)
```

```python
#Day9 project:Ingredient Checker

#Step1: Define the recipe ingredients
recipe_ingredients = {"flour","sugar","butter","eggs","milk"}

#Step2: Get user input for available ingredients
user_input = input("Enter the ingredients you have (separated by commas): ")
user_ingredients = set(user_input.split(","))

#Step3: Compare the Ingredients
missing_ingredients = recipe_ingredients - user_ingredients
extra_ingredients = user_ingredients - recipe_ingredients

#Step4: Display Result
print("\n---- Ingredient Check Result ----")
if missing_ingredients:
    print(f"You are missing the following ingredients: {','.join(missing_ingredients)}")
else:
    print(f"You hanve all the ingredients.")

if extra_ingredients:
    print(f"Extra ingredients you have: {', '.join(extra_ingredients)}")
else:
    print("You hava all the ingredients needed.")
```
