```python
#List operation:List operations are actions you can perform on lists to manipulate, analyze, or transform data.

shopping_list = ["milk","bread","egg"]
shopping_list.append("apple") #for adding items
print(shopping_list)

shopping_list.remove("milk") #for removing items
print(shopping_list)

shopping_list.sort() #for sorting items
print(shopping_list)

shopping_list.reverse() #for reversing items
print(shopping_list)

shopping_list.insert(1,"orange") #for inserting items
print(shopping_list)

shopping_list.pop(2) #for removing items
print(shopping_list)

shopping_list.clear() #for clearing items
print(shopping_list)

shopping_list = ["milk","egg","bread"]
print(shopping_list[1]) #for getting index
print(shopping_list[-1]) #for getting last index

#loop in listing:
for item in shopping_list: #for looping
    print(item)

for index,items in enumerate(shopping_list):
    print(f"{index + 1}. {items}")
```
