```python
# While Loop : repeats a block of code as long as the given condition becomes true.

count = 7
while count > 0:
  print(count)
  count -= 1

```

```python
count = 0
while count <= 7:
  print(count)
  count += 1
```

```python
#simply concepts of (increment and decrement)
x = 10

# Method 1: Full way
x = x + 1   # x = 11
x = x - 1   # x = 10

# Method 2: Shorter way (MOST COMMON)
x += 1      # x = 11 (increment)
x -= 1      # x = 10 (decrement)

# Method 3: Multiply (increase by more)
x *= 2      # x = 20 (double)
x /= 2      # x = 10 (half)

# Method 4: Add any number
x += 5      # x = 15 (increase by 5)
x -= 3      # x = 12 (decrease by 3)
```

```python
#Day5 project; Countdown Timer

import time
#step1: Get user input for countdown start
duration = int(input("Enter the duration in Second!"))

#Step2: countdoun using a while loop
print("\n--- Countdown Begins---")
while  duration > 0:
  print(duration)
  time.sleep(1)
  duration -= 1
print("Countdown Complete!")
```
