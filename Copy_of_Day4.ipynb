```python
#Nested if-else statement, one (if else )statement is inside in other (if else) statement, like a one decision inside the others decision
marks = 77

if marks >= 33:
  print("You are Passed")
if marks >=90:
  print("Grade: A (Excellent)" )
elif marks >=70:
  print("Grade: B (Very Good)" )
elif marks >=60:
  print("Grade: C+ (Good)" )
elif marks >=40:
  print("Grade: C (Average)")
elif marks >=33:
  print("Grade: D (Need to improve)")
else:
  print("Grade: F (Fail)")


```
