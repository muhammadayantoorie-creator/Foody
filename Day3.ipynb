```python
#Day3 project, simple calculator

#step1, Get user input for two numbers
num1 = float(input("Enter the first number:"))
num2 = float(input("Enter the second number:"))

#step2,perform arthimatic operation
addition = num1 + num2
substraction = num1 - num2
multipication= num1 * num2
division = num1 / num2 if num2 !=0 else "Number can not divide by zero"

#step3,, Display the result
print("\n-- Calculator Result--")
print(f"Addition: {num1} + {num2} = {addition}")
print(f"Substraction: {num1} - {num2} = {substraction}")
print(f"Multipication: {num1} * {num2} = {multipication}")
print(f"Division: {num1} / {num2} = {division}")

```
